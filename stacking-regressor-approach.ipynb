```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, LinearRegression
from sklearn.ensemble import StackingRegressor
```

```python
train = pd.read_csv("/kaggle/input/playground-series-s4e5/train.csv")
test = pd.read_csv("/kaggle/input/playground-series-s4e5/test.csv")
```

```python
test= test.drop("id", axis =1)
```

```python
x= train.drop(["FloodProbability","id"], axis=1)
y = train[["FloodProbability"]]
```

```python
x.head()
```

```python
y.head()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.20, random_state=1)
```

```python
ridge_model = Ridge(alpha=.001,  tol=0.1) # Tune alpha parameter
```

```python
re = LinearRegression()
```

```python
base_regressors = [
    ('rg', ridge_model),
    ('lr', LinearRegression())
]
```

```python
meta_regressor = LinearRegression()
```

```python
stacking_regressor = StackingRegressor(estimators=base_regressors, final_estimator=meta_regressor)
```

```python
stacking_regressor.fit(xtrain,ytrain)

# Evaluate on test set
score = stacking_regressor.score(xtest, ytest)
print("Stacking Regressor R2 Score:", score)
```

```python
pred = stacking_regressor.predict(test)

# Create DataFrame with predictions and columns "id" and "FloodProbability"
submission = pd.DataFrame({'id': test.index, 'FloodProbability': pred})

# Save predictions to a CSV file
submission.to_csv('submission.csv', index=False)
```
